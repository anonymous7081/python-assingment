### Exercise 2: Creating DataFrames from Series

1. Create two Series: one containing student names and another containing their ages. Combine them into a single DataFrame.
2. Create three Series: 'Product Name', 'Price', and 'Quantity'. Combine them into one DataFrame and display it.

```python
#answer here
Student_name = pd.Series (['Aditya','Manish','Avinash','Akhilesh', 'Abhishek'])
Student_ages = pd.Series ([25,23,25,26,22])
df= pd.concat([ Student_name, Student_ages], axis=1)
df.columns = (['Student_name','Student_ages'])
df
```

```python

Product_Name = pd.Series(['Apple','raw mango','Banana','Guwawa','Pineapple'])
Price = pd.Series([120,80,60,80,140])
Quantity = pd.Series([2,8,6,12,2])
df = pd.concat([Product_Name,Price,Quantity], axis= 1)
df
```
